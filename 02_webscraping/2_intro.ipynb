W internecie isnieją strony stworzone specjalnie w celach ćwiczenie scrappingu. Jedną z nich jest https://www.scrapethissite.com/. Zacznijmy od niej.

```python
import requests

url = "https://www.scrapethissite.com/pages/simple/"
response = requests.get(url)
response
```

```python
response.text
```

```python
content = response.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(content, 'html.parser')
soup
```

```python
countries = soup.find_all(class_="country-name")

capitals = soup.find_all(class_="country-capital")
populations = soup.find_all(class_="country-population")
areas = soup.find_all(class_="country-area")
```

```python
for idx in range(len(countries)):
    country = countries[idx].text.strip()
    capital = capitals[idx].text.strip()
    population = populations[idx].text.strip()
    area = areas[idx].text.strip()
    
    print(f"{country}: [{capital}, {population}, {area}]")
```

```python
def get_soup(url):
    """Takes URL and returns a soup object."""

    response = requests.get(url)
    if response.status_code == 200:
        return BeautifulSoup(response.text, 'html.parser')
    else: 
        return None
```

```python
def get_countries(soup):
    data = []
    countries = soup.find_all('div', class_='country')
    for country in countries:
        country_details = get_details(country)
        data.append(country_details)
    
    return data
```

```python
def get_details(country):
    name = country.find(class_="country-name").text.strip()
    captial = country.find(class_="country-capital").text.strip()
    population = country.find(class_="country-population").text.strip()
    area = country.find(class_="country-area").text.strip()

    return (name, capital, population, area)
```

```python
soup = get_soup("https://www.scrapethissite.com/pages/simple/")
if soup:
    countries = get_countries(soup)

countries[:5]
```
