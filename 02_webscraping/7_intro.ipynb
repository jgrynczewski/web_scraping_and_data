Innym ciekawym przykładem jest strona https://www.imdb.com/chart/top/ popularnego portalu IMDb (Internet Movie Database). Na stronie wyświetlane jest 250 najpopularniejszych filmów. Tylko, że początkowo wyświetla się 25, a kolejne są dynamicznie dodawane w tle przez javascript, w czasie kiedy pierwsze 25 pozycji jest już wyświetlonych i dostępnych dla użytkownika.

Tym samym uderzając na portal w tradycyjny dostaniemy zwrotkę tylko z tymi 25 filmami.

```python
import requests

headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/130.0.0.0 Safari/537.36 Edg/130.0.0.0"
}  # portal sprawdza nagłówek
url = "https://www.imdb.com/chart/top/"

res = requests.get(url, headers=headers)
content = res.text

from bs4 import BeautifulSoup

soup = BeautifulSoup(content, 'html.parser')

movies = soup.select(".ipc-metadata-list-summary-item .ipc-title__text")
for movie in movies:
    print(movie.text)
```

W jaki sposób możemy uwzględnić filmy, które są doładowywany w późniejszym czasie ?

WebDriver Selenium ma wśród swoich pól metodę `implicitly_wait`, która nakazuje silnikowi w przypadku wywołania metody wyszującej jakiś element (find, select, ...) odczekać ilość sekund wskazaną w metodzie `implicit_wait` zanim zacznie wyszukiwanie.

```python
from selenium import webdriver
from selenium.webdriver.common.by import By

driver = webdriver.Edge()
driver.implicitly_wait(20)
driver.get("https://www.imdb.com/chart/top/")
movies = driver.find_elements(By.CSS_SELECTOR, ".ipc-metadata-list-summary-item .ipc-title__text")

for movie in movies:
    print(movie.text)
```
