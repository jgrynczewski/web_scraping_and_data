### Dwa, gwne tryby dziaania

Po Jupyter Lab mona porusza si za pomoc myszki, albo za pomoc skrtw klawiszowych. W przypadku uzywania skrtw klawiszowych Jupyter Lab posiada dwa tryby:
- tryb edycji (edit mode)
- tryb polece (command mode)

Kady z trybw ma swj zestaw skrtw klawiszowych.

#### Przykady skrtw klawiszowych w trybie polecenia

- `Ctrl+Enter` - wykonanie komrki
- `Shift+Enter` - wykonanie komrki i przejcie do kolejnej lub jeeli jestemy na samym kocu, stworzenie kolejnej.
- `b` (below) - utwrz poniej
- `a` (above) - utwrz powyej
- `dd` (delete) - usuwa zaznaczon komrk
