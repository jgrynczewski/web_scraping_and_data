## Wstęp

Projekt Jupyter (https://jupyter.org/) jest udaną próbą stworzenia środowiska do interaktywnej pracy z kodem. Zawiera dwa flagowe produkty - **Jupyter Notebook** oraz **Jupyter Lab**. Oba pozwalają na interaktywne uruchamianie wybranych fragmentów kodu i uzupełniania kodu o wysokiej jakości notatki włączając w to możliwością umieszczania w notatkach wykresów, ilustracji, dźwięków i dowolnego rodzaju plików. W wybranych przypadkach może być też wygodnym narzędziem do prezentowania wynikow. Będziemy go używali głównie przez wzgląd na jego walory edukacyjne. Dzięki niemu nasz kod będzie bogato opatrzony notatkami.

Projekt Jupyter wyrósł z projektu [IPython](https://ipython.org/) - Interactive Python, który nazywany jest czasami "interpreterem pythona na sterydach". Twórcą obu projektów jest [Fernando Pérez](https://en.wikipedia.org/wiki/Fernando_P%C3%A9rez_(software_developer))

### Dwa, główne tryby działania

Po Jupyter Lab można poruszać się za pomocą myszki, albo za pomocą skrótów klawiszowych. W przypadku uzywania skrótów klawiszowych Jupyter Lab posiada dwa tryby:
- tryb edycji (edit mode)
- tryb poleceń (command mode)

Każdy z trybów ma swój zestaw skrótów klawiszowych.

#### Przykłady skrótów klawiszowych w trybie polecenia

- `Ctrl+Enter` - wykonanie komórki
- `Shift+Enter` - wykonanie komórki i przejście do kolejnej lub jeżeli jesteśmy na samym końcu, stworzenie kolejnej.
- `b` (below) - utwórz poniżej
- `a` (above) - utwórz powyżej
- `dd` (delete) - usuwa zaznaczoną komórkę
